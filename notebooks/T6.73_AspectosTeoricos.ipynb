# Estudios de Correlacion

```python
import pandas as pd
import numpy as np
```

```python
data_ads = pd.read_csv('../data/ads/Advertising.csv')
data_ads
```

```python
corr_TV= sum((data_ads["TV"] - np.mean(data_ads["TV"])) * (data_ads["Sales"] - np.mean(data_ads["Sales"])))/ \
                     np.sqrt(sum(
                                    (data_ads["TV"] - np.mean(data_ads["TV"]))**2
                                ) *
                             sum (
                                    (data_ads["Sales"] - np.mean(data_ads["Sales"]))**2
                                  )
                     )
```

```python
corr_TV
```

```python
def correlation_coef (dataFrame, var1, var2) -> float:
    return sum((dataFrame[var1] - np.mean(dataFrame[var1])) * (dataFrame[var2] - np.mean(dataFrame[var2])))/ \
                     np.sqrt(sum(
                                    (dataFrame[var1] - np.mean(dataFrame[var1]))**2
                                ) *
                             sum (
                                    (dataFrame[var2] - np.mean(dataFrame[var2]))**2
                                  )
                     )
```

```python
print (correlation_coef(data_ads, "TV", "Sales"))
print (correlation_coef(data_ads, "Radio", "Sales"))
print (correlation_coef(data_ads, "Newspaper", "Sales"))
```

```python
for x in data_ads.columns.values:
    for y in data_ads.columns.values:
        print (f'La correlación entre {x} e {y} es: {correlation_coef(data_ads, x, y)}')
```
